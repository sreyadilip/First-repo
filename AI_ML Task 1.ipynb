```python
m=np.random.randint(0,100,(50))
bins=np.arange(0,100,5)
plt.hist(m,bins=bins)
plt.xticks(np.arange(0,100,5))
```

```python
classes=['phy','chem','maths','science','sst']
marks=[89,45,78,23,90]
plt.pie(marks,labels=classes)
```

```python
colors=['red','blue','green','yellow','violet']
explode_values=[0.1,0,0,0.2,0]
plt.pie(marks,labels=classes,colors=colors,autopct='%0.1f%%',explode=explode_values,radius=1.2)
```

```python
textprops={'fontsize':15,'color':'green'}
wedgeprops={'linewidth':3,'linestyle':'--','edgecolor':'white'}
plt.pie(marks,labels=classes,colors=colors,autopct='%0.1f%%',explode=explode_values,radius=1.6,textprops=textprops,wedgeprops=wedgeprops)
plt.legend()
```

```python
study_hours=[2,3,4,5,6,7,7,8,9]
marks=[10,20,30,40,50,60,70,80,90]
plt.scatter(study_hours,marks)
```

```python
plt.plot(study_hours,marks)
```

```python
plt.hist(marks)
```

```python
plt.plot(study_hours,marks,'r-')
plt.plot(study_hours,marks,'go')
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.scatter(study_hours,marks)
plt.subplot(2,2,2)
plt.plot(study_hours,marks)
plt.subplot(2,2,3)
plt.hist(marks)
plt.subplot(2,2,4)
plt.plot(study_hours,marks,'r-')
plt.plot(study_hours,marks,'go')

plt.savefig('/content/subplot',facecolor='k')
```
